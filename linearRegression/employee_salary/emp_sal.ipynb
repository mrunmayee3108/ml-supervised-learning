feature engineering 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()    # converting categorical columns into numerical columns. algo wont understand text.
```

```python
df1['Gender_encode'] = le.fit_transform(df1['Gender'])
df1['Degree_encode'] = le.fit_transform(df1['Degree'])
df1['Job_Title_encode'] = le.fit_transform(df1['Job_Title'])
```

```python
df1.head()

# male = 1, female = 0
# bachelor = 0, masters = 1, phd = 2
```

```python
# bring the age and experience columns to same scale.
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df1['Age_scaled'] = std_scaler.fit_transform(df1[['Age']])
df1['Experience_years_scaled'] = std_scaler.fit_transform(df1[['Experience_years']])
```

```python
df1.head()   
```

```python
# independent and dependent variables

x = df1[['Age_scaled', 'Gender_encode', 'Degree_encode', 'Experience_years_scaled']] # ind
y = df1['Salary'] # dep
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
# train the model
lr.fit(x_train, y_train)
```

predictions

```python
y_pred_lr = lr.predict(x_test)
```

```python
y_pred_lr
```

```python
dff = pd.DataFrame({
    'y_actual': y_test, 
    'y_predicted': y_pred_lr
})
dff['error'] = dff['y_actual']-dff['y_predicted']
dff['abs_error'] = abs(dff['error'])
mean_abs_err = dff['abs_error'].mean()
mean_abs_err
```

```python
dff
```

model evaluation

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
r2score = r2_score(y_test, y_pred_lr)
```

```python
print(f"Accuracy of the model: {round(r2score, 4)*100}%")   
# The model explains 89.13% of the variance in salary.
# strong linear relationship -> selected features are doing a good job
```

```python
print(f"Mean absolute error: {round(mean_absolute_error(y_test, y_pred_lr), 2)}")
# On average, salary predictions are off by about 10.5k.
```

```python
mse = mean_squared_error(y_test, y_pred_lr)
```

```python
print(f"Mean squared error: {round(mse, 2)}")
```

```python
print("Root mean squared error (RMSE):", mse**0.5)
```

The linear regression model explains approximately 89% of the variance in employee salaries. The model achieves a mean absolute error of 10.5k and an RMSE of 14.3k, indicating reasonably accurate predictions with occasional larger errors. Overall, the model demonstrates strong predictive performance
