```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)   
df.head()
```

```python
df.shape
```

```python
df['target'] = iris.target
df.head()

# 0 - setosa
# 1 - versicolor    
# 2 - virginica
```

```python
df[df['target'] == 0]  # setosa
```

```python
df[df['target'] == 1]  # versicolor
```

```python
df[df['target'] == 2]  # virginica
```

```python
df_setosa = df[:50]
df_versicolor = df[50:100]
df_virginica = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df_setosa['sepal length (cm)'], df_setosa['sepal width (cm)'], color='red', label='setosa')
plt.scatter(df_versicolor['sepal length (cm)'], df_versicolor['sepal width (cm)'], color='green', label='versicolor')
plt.scatter(df_virginica['sepal length (cm)'], df_virginica['sepal width (cm)'], color='blue', label='virginica')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('target', axis='columns')
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1, stratify = y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 16)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
dff = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred
})
dff
```

```python
knn.score(X_test, y_test) # accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
