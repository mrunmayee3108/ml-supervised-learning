```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('species', axis = 1)
y = df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=7)
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
model_svc = SVC(gamma='auto')
```

```python
model_svc.fit(X_train, y_train)
model_svc.score(X_test, y_test)
```

now using grid search cv

```python
from sklearn.model_selection import GridSearchCV
classifier = GridSearchCV(model_svc, {
    'C': [1, 10, 20, 30, 40, 50],
    'kernel': ['rbf', 'linear']
}, cv=5, return_train_score=False)
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
results
```

```python
results[['param_C', 'param_kernel', 'mean_test_score']]
```

ensemble learning-

stacking

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size = 0.3, random_state = 42, stratify=y_encoded)

# setosa- 0, versicolor- 1, virginica- 2
```

```python
base_models = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svm', SVC(probability=True, gamma='auto', C=10, kernel='rbf', random_state=42)),
    ('lr', LogisticRegression(random_state=42, max_iter=1000))
]
```

```python
meta_model = RandomizedSearchCV(model_knn, {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],
    'weights': ['uniform', 'distance'], 
    'metric': ['euclidean', 'manhattan']
}, n_iter = 4, cv=5, return_train_score=False)
meta_model.fit(X_train, y_train)  # n = 11 is the best
```

```python
y_pred_meta = meta_model.predict(X_test)
accuracy_meta = accuracy_score(y_test, y_pred_meta)
print(f"Meta model accuracy: {accuracy_meta:.4f}")
```

```python
stacking_classifier = StackingClassifier(estimators=base_models, final_estimator=meta_model, cv=5)
```

```python
stacking_classifier.fit(X_train, y_train)
```

```python
y_pred = stacking_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

bagging

random forest

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100,  # no. of trees
                                  max_depth=None,  # max depth of each tree. here None. we let them grow fully.
                                  random_state=42)
```

```python
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_rf
```

boosting

adaboost
- use when underfitting
- i.e when high bias, high variance.. when the model doesnt perform well on both training n testing data.

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
```

```python
ada_model = AdaBoostClassifier(n_estimators=100, random_state=42)
ada_model.fit(X_train, y_train)
```

```python
y_pred_ada = ada_model.predict(X_test)
accuracy_ada = accuracy_score(y_test, y_pred_ada)
accuracy_ada
```

gradient boosting

```python
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb_model.fit(X_train, y_train)
```

```python
y_pred_gb = gb_model.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
accuracy_gb
```

```python
xgb_model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth = 3, random_state=42, use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)   
accuracy_xgb
```

use bagging for underfitting, boosting for overfitting
